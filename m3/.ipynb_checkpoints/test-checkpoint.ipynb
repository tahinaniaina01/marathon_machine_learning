# Customer Segmentation – Online Retail Dataset

**Steps covered:**
1. Exploratory Data Analysis (EDA)
2. Feature Engineering (RFM)
3. Clustering – K-Means (modelling + interpretation)
4. Dimensionality Reduction after clustering – PCA, t-SNE, DBSCAN (plots + interpretation)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import seaborn as sns
from datetime import datetime

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, DBSCAN
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.metrics import silhouette_score, silhouette_samples

import warnings
warnings.filterwarnings('ignore')

sns.set_theme(style='whitegrid', palette='tab10')
plt.rcParams['figure.dpi'] = 110
print("Libraries loaded ✓")
```

---
## 1. Exploratory Data Analysis (EDA)

```python
# ── Load Data ──────────────────────────────────────────────────────────────────
df = pd.read_excel('data.xlsx')
print(f"Shape: {df.shape}")
df.head()
```

```python
# ── 1.1 Basic info ────────────────────────────────────────────────────────────
print("=== dtypes ===")
print(df.dtypes)
print("\n=== describe (numeric) ===")
df.describe()
```

```python
# ── 1.2 Missing values ────────────────────────────────────────────────────────
missing = df.isnull().sum()
missing_pct = (missing / len(df) * 100).round(2)
pd.DataFrame({'Missing': missing, 'Pct (%)': missing_pct}).sort_values('Missing', ascending=False)
```

```python
# ── 1.3 Distributions of Quantity and UnitPrice ───────────────────────────────
fig, axes = plt.subplots(1, 2, figsize=(14, 4))

# Clip outliers for readability
q_clip = df['Quantity'].clip(lower=0, upper=df['Quantity'].quantile(0.99))
p_clip = df['UnitPrice'].clip(lower=0, upper=df['UnitPrice'].quantile(0.99))

sns.histplot(q_clip, bins=60, ax=axes[0], color='steelblue', kde=True)
axes[0].set_title('Quantity Distribution (0–99th pctile)')
axes[0].set_xlabel('Quantity')

sns.histplot(p_clip, bins=60, ax=axes[1], color='coral', kde=True)
axes[1].set_title('Unit Price Distribution (0–99th pctile)')
axes[1].set_xlabel('Unit Price (£)')

plt.tight_layout()
plt.show()
print("\nQuantity < 0 (returns):", (df['Quantity'] < 0).sum())
print("UnitPrice <= 0:", (df['UnitPrice'] <= 0).sum())
```

```python
# ── 1.4 Top 10 countries by revenue ──────────────────────────────────────────
df_valid = df[(df['Quantity'] > 0) & (df['UnitPrice'] > 0)].copy()
df_valid['Revenue'] = df_valid['Quantity'] * df_valid['UnitPrice']

top_countries = (df_valid.groupby('Country')['Revenue']
                 .sum()
                 .sort_values(ascending=False)
                 .head(10))

plt.figure(figsize=(12, 4))
sns.barplot(x=top_countries.values, y=top_countries.index, palette='Blues_r')
plt.title('Top 10 Countries by Revenue')
plt.xlabel('Total Revenue (£)')
plt.tight_layout()
plt.show()
```

```python
# ── 1.5 Monthly revenue trend ────────────────────────────────────────────────
df_valid['YearMonth'] = df_valid['InvoiceDate'].dt.to_period('M')
monthly = df_valid.groupby('YearMonth')['Revenue'].sum()

plt.figure(figsize=(13, 4))
monthly.plot(kind='bar', color='teal', edgecolor='white')
plt.title('Monthly Revenue Trend')
plt.xlabel('Month')
plt.ylabel('Revenue (£)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# ── 1.6 Orders per customer distribution ─────────────────────────────────────
orders_per_customer = df_valid.groupby('CustomerID')['InvoiceNo'].nunique()

plt.figure(figsize=(10, 4))
sns.histplot(orders_per_customer.clip(upper=orders_per_customer.quantile(0.99)),
             bins=50, color='mediumpurple', kde=True)
plt.title('Number of Orders per Customer (up to 99th percentile)')
plt.xlabel('# Distinct Invoices')
plt.tight_layout()
plt.show()
print(f"Total unique customers with CustomerID: {df_valid['CustomerID'].nunique()}")
```

---
## 2. Feature Engineering – RFM

RFM (Recency, Frequency, Monetary) is the standard customer segmentation framework:
- **Recency** – days since last purchase (lower = more recent = better)
- **Frequency** – number of distinct invoices
- **Monetary** – total spend (£)

```python
# ── Build RFM table ────────────────────────────────────────────────────────────
# Keep only transactions with a known customer
rfm_data = df_valid.dropna(subset=['CustomerID']).copy()
rfm_data['CustomerID'] = rfm_data['CustomerID'].astype(int)

snapshot_date = rfm_data['InvoiceDate'].max() + pd.Timedelta(days=1)

rfm = rfm_data.groupby('CustomerID').agg(
    Recency   = ('InvoiceDate', lambda x: (snapshot_date - x.max()).days),
    Frequency = ('InvoiceNo',   'nunique'),
    Monetary  = ('Revenue',     'sum')
).reset_index()

print(f"RFM table shape: {rfm.shape}")
rfm.describe()
```

```python
# ── RFM Distributions ─────────────────────────────────────────────────────────
fig, axes = plt.subplots(1, 3, figsize=(16, 4))
colors = ['#4C72B0', '#DD8452', '#55A868']

for ax, col, c in zip(axes, ['Recency', 'Frequency', 'Monetary'], colors):
    clip_val = rfm[col].quantile(0.99)
    sns.histplot(rfm[col].clip(upper=clip_val), bins=50, ax=ax, color=c, kde=True)
    ax.set_title(f'{col} Distribution')
    ax.set_xlabel(col)

plt.suptitle('RFM Distributions (clipped at 99th percentile)', y=1.02, fontsize=13)
plt.tight_layout()
plt.show()
```

```python
# ── Log-transform + Standard Scale ────────────────────────────────────────────
rfm_log = rfm[['Recency', 'Frequency', 'Monetary']].copy()
# Log1p to reduce right skew
rfm_log = np.log1p(rfm_log)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(rfm_log)
X_scaled = pd.DataFrame(X_scaled, columns=['Recency', 'Frequency', 'Monetary'])

print("Scaled RFM – stats after transformation:")
X_scaled.describe().round(3)
```

---
## 3. Clustering – K-Means

We use the **Elbow method** (inertia) and the **Silhouette score** to choose the optimal number of clusters.

```python
# ── 3.1 Elbow + Silhouette ────────────────────────────────────────────────────
inertia   = []
sil_scores = []
K_range = range(2, 11)

for k in K_range:
    km = KMeans(n_clusters=k, init='k-means++', n_init=10, random_state=42)
    km.fit(X_scaled)
    inertia.append(km.inertia_)
    sil_scores.append(silhouette_score(X_scaled, km.labels_))
    print(f"k={k}  inertia={km.inertia_:.1f}  silhouette={sil_scores[-1]:.4f}")

fig, axes = plt.subplots(1, 2, figsize=(14, 4))

axes[0].plot(list(K_range), inertia, 'o-', color='steelblue')
axes[0].set_title('Elbow Method – Inertia')
axes[0].set_xlabel('Number of clusters (k)')
axes[0].set_ylabel('Inertia')

axes[1].plot(list(K_range), sil_scores, 's-', color='coral')
axes[1].set_title('Silhouette Score')
axes[1].set_xlabel('Number of clusters (k)')
axes[1].set_ylabel('Silhouette score')

plt.tight_layout()
plt.show()
```

```python
# ── 3.2 Final K-Means model (k=4) ────────────────────────────────────────────
OPTIMAL_K = 4   # adjust if elbow/silhouette suggests otherwise

kmeans = KMeans(n_clusters=OPTIMAL_K, init='k-means++', n_init=10, random_state=42)
rfm['Cluster'] = kmeans.fit_predict(X_scaled)

print(f"Cluster sizes:")
print(rfm['Cluster'].value_counts().sort_index())
print(f"\nOverall silhouette: {silhouette_score(X_scaled, rfm['Cluster']):.4f}")
```

```python
# ── 3.3 Cluster interpretation ───────────────────────────────────────────────
cluster_profile = rfm.groupby('Cluster')[['Recency', 'Frequency', 'Monetary']].mean().round(1)
cluster_profile['Size'] = rfm['Cluster'].value_counts().sort_index()
print("=== Cluster Profiles (original RFM scale) ===")
display(cluster_profile)

# Snake plot
rfm_norm = rfm[['Recency', 'Frequency', 'Monetary', 'Cluster']].copy()
for col in ['Recency', 'Frequency', 'Monetary']:
    rfm_norm[col] = (rfm_norm[col] - rfm_norm[col].min()) / (rfm_norm[col].max() - rfm_norm[col].min())

snake = rfm_norm.groupby('Cluster')[['Recency', 'Frequency', 'Monetary']].mean()

plt.figure(figsize=(9, 4))
for cluster_id in snake.index:
    plt.plot(snake.columns, snake.loc[cluster_id], marker='o', label=f'Cluster {cluster_id}', linewidth=2)
plt.title('Snake Plot – Normalised RFM by Cluster')
plt.ylabel('Normalised value (0–1)')
plt.legend(title='Cluster')
plt.tight_layout()
plt.show()
```

```python
# ── 3.4 Silhouette plot per cluster ──────────────────────────────────────────
fig, ax = plt.subplots(figsize=(9, 5))
y_lower = 10
sil_vals = silhouette_samples(X_scaled, rfm['Cluster'])
colors_map = cm.tab10(np.linspace(0, 0.5, OPTIMAL_K))

for i in range(OPTIMAL_K):
    cluster_sil = np.sort(sil_vals[rfm['Cluster'] == i])
    size_i = len(cluster_sil)
    y_upper = y_lower + size_i
    ax.fill_betweenx(np.arange(y_lower, y_upper), 0, cluster_sil,
                     facecolor=colors_map[i], edgecolor='none', alpha=0.85)
    ax.text(-0.05, y_lower + 0.5 * size_i, str(i))
    y_lower = y_upper + 10

ax.axvline(x=silhouette_score(X_scaled, rfm['Cluster']), color='red', linestyle='--',
           label='Average silhouette')
ax.set_title('Silhouette Plot per Cluster')
ax.set_xlabel('Silhouette coefficient')
ax.set_ylabel('Cluster label')
ax.legend()
plt.tight_layout()
plt.show()
```

## 4. Dimensionality Reduction after Clustering

We explore three approaches to visualise and validate the clusters found by K-Means:

| Method | Type | Purpose |
|--------|------|---------|
| **PCA** | Linear | Fast, preserves global variance, interpretable components |
| **t-SNE** | Non-linear | Preserves local neighbourhood structure, great for visualisation |
| **DBSCAN** | Density-based clustering | Alternative clustering approach – validates K-Means; identifies outliers |

```python
# ── 4.1 PCA – explained variance ─────────────────────────────────────────────
pca_full = PCA(random_state=42)
pca_full.fit(X_scaled)

cumvar = np.cumsum(pca_full.explained_variance_ratio_)

plt.figure(figsize=(7, 4))
plt.bar(range(1, len(cumvar)+1), pca_full.explained_variance_ratio_, label='Individual', color='steelblue')
plt.step(range(1, len(cumvar)+1), cumvar, where='mid', label='Cumulative', color='red')
plt.axhline(0.95, color='gray', linestyle='--', label='95% threshold')
plt.title('PCA – Explained Variance')
plt.xlabel('Principal Component')
plt.ylabel('Variance Ratio')
plt.legend()
plt.tight_layout()
plt.show()

print("Explained variance per PC:", pca_full.explained_variance_ratio_.round(4))
```

```python
# ── 4.2 PCA 2D scatter coloured by K-Means cluster ───────────────────────────
pca2 = PCA(n_components=2, random_state=42)
X_pca = pca2.fit_transform(X_scaled)

plt.figure(figsize=(9, 6))
scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1],
                      c=rfm['Cluster'], cmap='tab10', alpha=0.6, s=15)
plt.colorbar(scatter, label='K-Means Cluster')
plt.title(f'PCA (2D) – K-Means Clusters\n'
          f'PC1={pca2.explained_variance_ratio_[0]:.2%}  '
          f'PC2={pca2.explained_variance_ratio_[1]:.2%}')
plt.xlabel('PC 1')
plt.ylabel('PC 2')

# Overlay cluster centroids projected into PCA space
centroids_pca = pca2.transform(kmeans.cluster_centers_)
plt.scatter(centroids_pca[:, 0], centroids_pca[:, 1],
            marker='X', s=200, c='black', label='Centroids', zorder=5)
plt.legend()
plt.tight_layout()
plt.show()

# Loadings
loadings = pd.DataFrame(pca2.components_.T,
                         columns=['PC1', 'PC2'],
                         index=['Recency', 'Frequency', 'Monetary'])
print("PCA Loadings:")
display(loadings)
```

```python
# ── 4.3 t-SNE 2D scatter ─────────────────────────────────────────────────────
# t-SNE is slower; subsample if dataset is large
np.random.seed(42)
N_TSNE = min(5000, len(X_scaled))
idx = np.random.choice(len(X_scaled), N_TSNE, replace=False)

X_tsne_input = X_scaled.values[idx]
labels_tsne  = rfm['Cluster'].values[idx]

tsne = TSNE(n_components=2, perplexity=40, max_iter=1000,
            learning_rate='auto', init='pca', random_state=42)
X_tsne = tsne.fit_transform(X_tsne_input)

plt.figure(figsize=(9, 6))
scatter = plt.scatter(X_tsne[:, 0], X_tsne[:, 1],
                      c=labels_tsne, cmap='tab10', alpha=0.6, s=15)
plt.colorbar(scatter, label='K-Means Cluster')
plt.title(f't-SNE (2D) – K-Means Clusters  (n={N_TSNE})')
plt.xlabel('t-SNE 1')
plt.ylabel('t-SNE 2')
plt.tight_layout()
plt.show()
```

```python
# ── 4.4 DBSCAN on PCA-reduced data ───────────────────────────────────────────
# Run DBSCAN in the 2-D PCA space for fair comparison with the scatter plots

# Tune eps using k-NN distance plot (k = min_samples - 1)
from sklearn.neighbors import NearestNeighbors

k_nn = 4
nbrs = NearestNeighbors(n_neighbors=k_nn).fit(X_pca)
distances, _ = nbrs.kneighbors(X_pca)
k_dist = np.sort(distances[:, -1])

plt.figure(figsize=(9, 4))
plt.plot(k_dist, color='darkgreen')
plt.title(f'k-NN Distance Plot (k={k_nn}) – Elbow ≈ eps for DBSCAN')
plt.xlabel('Points sorted by distance')
plt.ylabel(f'{k_nn}-NN distance')
plt.tight_layout()
plt.show()
```

```python
# ── 4.5 DBSCAN fit ────────────────────────────────────────────────────────────
# eps chosen from elbow of k-NN distance plot above
DBSCAN_EPS      = 0.3   # adjust after inspecting the elbow plot
DBSCAN_MIN_SAMP = 10

db = DBSCAN(eps=DBSCAN_EPS, min_samples=DBSCAN_MIN_SAMP)
db_labels = db.fit_predict(X_pca)

n_clusters_db = len(set(db_labels)) - (1 if -1 in db_labels else 0)
n_noise       = (db_labels == -1).sum()

print(f"DBSCAN → {n_clusters_db} cluster(s),  {n_noise} noise points ({n_noise/len(db_labels)*100:.1f}%)")
if n_clusters_db > 1:
    mask = db_labels != -1
    print(f"Silhouette (non-noise): {silhouette_score(X_pca[mask], db_labels[mask]):.4f}")

# Side-by-side comparison: K-Means vs DBSCAN in PCA space
fig, axes = plt.subplots(1, 2, figsize=(16, 6), sharex=True, sharey=True)

# K-Means
sc1 = axes[0].scatter(X_pca[:, 0], X_pca[:, 1],
                       c=rfm['Cluster'], cmap='tab10', alpha=0.5, s=10)
axes[0].set_title('K-Means clusters in PCA space')
axes[0].set_xlabel('PC 1'); axes[0].set_ylabel('PC 2')
plt.colorbar(sc1, ax=axes[0], label='K-Means cluster')

# DBSCAN (-1 = noise in grey)
db_cmap = cm.get_cmap('tab10', n_clusters_db)
colors_db = [db_cmap(l) if l != -1 else (0.5, 0.5, 0.5, 0.4) for l in db_labels]
axes[1].scatter(X_pca[:, 0], X_pca[:, 1],
                c=colors_db, alpha=0.5, s=10)
axes[1].set_title(f'DBSCAN clusters (eps={DBSCAN_EPS}) in PCA space\n'
                  f'(grey = noise)')
axes[1].set_xlabel('PC 1')

plt.suptitle('K-Means vs DBSCAN – Cluster Comparison', fontsize=13)
plt.tight_layout()
plt.show()
```

## 5. Summary & Interpretation

### 5.1 K-Means Cluster Profiles (k=4, silhouette=0.34)

| Cluster | Recency (days) | Frequency | Monetary (£) | Size | Business Label |
|---------|---------------|-----------|--------------|------|----------------|
| **0** | 18 (very recent) | 2.1 (low) | 552 (low) | 837 | **New / Recent Customers** – bought recently but infrequently; nurture to loyalty |
| **1** | 12 (very recent) | 13.7 (high) | 8 074 (high) | 716 | **Champions** – best customers; highly engaged and high-value; retain & reward |
| **2** | 71 (moderate) | 4.1 (moderate) | 1 803 (moderate) | 1 173 | **Loyal / At-Risk** – regular buyers who have gone quiet; re-engage with promos |
| **3** | 183 (churned) | 1.3 (very low) | 344 (very low) | 1 612 | **Lost Customers** – long since last purchase; win-back campaign or deprioritise |

### 5.2 Dimensionality Reduction Comparison

| Method | Result | Interpretation |
|--------|--------|----------------|
| **PCA** | PC1 (75%) + PC2 (19%) = **94% variance**; 4 clusters visually separate along PC1 | PC1 is the "customer value" axis (Frequency ↑, Monetary ↑, Recency ↓ = more valuable). Best choice for communicating results. |
| **t-SNE** | Shows 4 clearly separated groups with tight cohesion | Confirms K-Means boundaries in the non-linear neighbourhood structure. Champions and Lost customers form the most distinct islands. |
| **DBSCAN** | 1 mega-cluster + 87 outliers (2%) | RFM-space is a continuous density gradient, not a multi-modal distribution. DBSCAN is not competitive here for segmentation. However, the **87 outlier points are genuine anomalies** (e.g. bulk B2B buyers) worth individual analysis. |

### 5.3 Chosen Approach

**PCA + K-Means (k=4)** is the selected final model:
- 2 PCA components explain **93.9%** of variance – minimal information loss with maximum interpretability.
- K-Means yields **4 actionable, well-separated business segments** confirmed by silhouette score (0.34) and the elbow at k=4.
- t-SNE independently validates the cluster structure in non-linear space.
- DBSCAN complements the analysis by flagging **87 high-value outlier customers** that deserve individual attention.

> **Recommended marketing actions by segment:**
> - Champions (C1): loyalty programmes, early-access offers
> - Loyal / At-Risk (C2): personalised re-engagement emails, special discounts
> - New Customers (C0): onboarding flows, first-repeat purchase incentives  
> - Lost (C3): final win-back campaign; if unresponsive, remove from active CRM
